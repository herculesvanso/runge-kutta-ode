```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from rungekutta import *
```

```python
def f(t,X):
    return np.array(
        [-20*X[0]+10*X[1]+100, 10*X[0]-20*X[1]]
    )
```

```python
a = 0.
b = 5.
X0 = np.array([0.,0.])
f = f
```

```python
tab1 = RK(a,b,X0,m=1000,f=f)
tab1
```

```python
tab2 = RK(a,b,X0,m=10,f=f)
tab2
```

```python
tab3 = RK(a,b=100,X0=X0,m=10**5,f=f)
tab3
```

```python
plt.plot(tab1.iloc[:,0],tab1.iloc[:,1],tab1.iloc[:,0],tab1.iloc[:,2])
plt.xlabel('Time (s)')
plt.ylabel('Current Intensity(A)')
plt.grid(True)
plt.show()
```

```python

plt.semilogy(tab2.iloc[:,0],np.abs(tab2.iloc[:,1]),tab2.iloc[:,0], np.abs (tab2.iloc[:,2]), 'ro')
plt.xlabel('Time (s) ')
plt.ylabel('Current Intensity(A)')
plt.grid(True)
plt.show()
```
