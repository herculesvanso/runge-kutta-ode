```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from rungekutta import rk
```

```python
def f(t,X):
    
    y1 = -20*X[0] + 10*X[1] + 100
    y2 = 10*X[0] - 20*X[1]

    return np.array([y1, y2])
```

```python
a = 0.
b = 5.
X0 = np.array([0., 0.])
f = f
```

```python
dict_tables = {'tab' + number: rk(a, b, X0, m= n_interval, f=f) for number, n_interval in zip("123", [10, 100, 1000])}
```
